```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("../Resources/schools_complete.csv")
student_data_to_load = Path("../Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Find number of schools
total_school = len(school_data_complete["school_name"].value_counts())
total_school
```

```python
#Total Students in data
total_students = len(school_data_complete["student_name"])
total_students
```

```python
#Total Budget for all schools
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Average math scores for all students
average_math = school_data_complete["maths_score"].mean()
average_math
```

```python
#Average reading scores for all students
average_read = school_data_complete["reading_score"].mean()
average_read
```

```python
#Percent of passing math students
pass_math = len(school_data_complete[school_data_complete["maths_score"]>= 50])/total_students
pass_math
```

```python
#Percent of passing reading students
pass_read = len(school_data_complete[school_data_complete["reading_score"]>= 50])/total_students
pass_read
```

```python
#Percent of overall passing
overall_pass = pass_math * pass_read
overall_pass
```

```python
#Summary into DataFrame
summary = {'Total Schools': [total_school],
                            'Total Students': [total_students],
                            'Total Budget': [total_budget],
                            'Average Maths Score': [average_math],
                            'Average Reading Score': [average_read],
                            '% Passing Maths': [pass_math*100],
                            '% Passing Reading': [pass_read*100],
                            '% Overall Passing': [overall_pass*100]}
area_summary = pd.DataFrame(summary)
area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Obtain school data from existing data
school_review = school_data[["school_name","type","size","budget"]]
school_review.head()
```

```python
school_index = school_review.set_index("school_name")
school_index.head()
```

```python
#Budget per student calculation
per_student = school_index["budget"]/school_index["size"]
school_index["Per Student Budget"] = per_student
school_index.head()
```

```python
school_grouped = school_data_complete.groupby('school_name')
school_grouped.head(20)
```

```python
#Looping through schools to calculate maths average score
for school_name, group in school_grouped:
    avg_math = school_grouped["maths_score"].mean()
avg_math
```

```python
#Adding to DataFrame
school_index["Average Maths Score"] = avg_math
school_index.head()
```

```python
#Looping through schools to calculate reading average score
for school_name, group in school_grouped:
    avg_read = school_grouped["reading_score"].mean()
avg_read
```

```python
#Adding to DataFrame
school_index["Average Reading Score"] = avg_read
school_index.head()
```

```python
#Finding number of students with passing math scores
math_pass = school_data_complete[school_data_complete['maths_score'] >= 50]
math_pass
```

```python
total_school = school_grouped["school_name"].value_counts()
total_school
```

```python
math_school_pass = math_pass["school_name"].value_counts()
math_school_pass
```

```python
per_math_pass = math_school_pass/total_school
per_math_pass
```

```python
#Adding to DataFrame
school_index["% Passing Maths"] = per_math_pass * 100
school_index.head()
```

```python
#Finding number of students with passing reading scores
read_pass = school_data_complete[school_data_complete['reading_score'] >= 50]
read_pass
```

```python
read_school_pass = read_pass["school_name"].value_counts()
read_school_pass
```

```python
per_read_pass = read_school_pass/total_school
per_read_pass
```

```python
#Adding to DataFrame
school_index["% Passing Reading"] = per_read_pass *100
school_index.head()
```

```python
#Adding to DataFrame
school_index["% Overall Passing"] = (per_read_pass * per_math_pass)*100
school_index
```

```python
per_school_summary = school_index.rename(columns={"school_name":"School Name",
                                                    "type":"School Type",
                                                    "size":"Total Students",
                                                   "budget":"Total School Budget"})
per_school_summary
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_school = school_index.sort_values("% Overall Passing", ascending=False)
top_school.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
bottom_school = school_index.sort_values("% Overall Passing", ascending=True)
bottom_school.head(5)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Find All year lvls in sheet
year_lvls = school_grouped['year'].unique()
year_lvls
```

```python
#Create Series for each year
year_9_math = school_data_complete[school_data_complete['year'] == 9]['maths_score']
year_10_math = school_data_complete[school_data_complete['year'] == 10]['maths_score']
year_11_math = school_data_complete[school_data_complete['year'] == 11]['maths_score']
year_12_math = school_data_complete[school_data_complete['year'] == 12]['maths_score']
```

```python
#Group Series by School for each year and calculate
school_yr_9_math = year_9_math.groupby(school_data_complete['school_name']).mean()
school_yr_10_math = year_10_math.groupby(school_data_complete['school_name']).mean()
school_yr_11_math = year_11_math.groupby(school_data_complete['school_name']).mean()
school_yr_12_math = year_12_math.groupby(school_data_complete['school_name']).mean()
```

```python
#Create DataFrame
maths_scores_by_year = pd.DataFrame({'Year 9': school_yr_9_math,
                                'Year 10': school_yr_10_math,
                                'Year 11': school_yr_11_math,
                                'Year 12': school_yr_12_math,})
maths_scores_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create Bins and labels
bins = [0,585,630,645,680]
spending_range = ["<$585","$585-630","$630-645","$645-680"]
```

```python
reindex = per_school_summary.reset_index()
reindex.head()
```

```python
spending = pd.DataFrame(reindex)
spending.head()
```

```python
#Adding Spending range column
spending["Spending Ranges (Per Student)"] = pd.cut(spending["Per Student Budget"], bins, labels=spending_range)
```

```python
#Setting Spending range as index
spending_indexed = spending.set_index('Spending Ranges (Per Student)')
spending_indexed.head()
```

```python
spending_grouped = spending_indexed[["Average Maths Score",
                              "Average Reading Score",
                              "% Passing Maths",
                              "% Passing Reading",
                              "% Overall Passing"]]
spending_grouped = pd.DataFrame(spending_grouped)
spending_grouped.head()
```

```python
average_math_spending = spending_grouped.groupby('Spending Ranges (Per Student)')['Average Maths Score'].mean()
average_math_spending
```

```python
average_read_spending = spending_grouped.groupby('Spending Ranges (Per Student)')['Average Reading Score'].mean()
average_read_spending
```

```python
avg_math_pass_spend = spending_grouped.groupby('Spending Ranges (Per Student)')['% Passing Maths'].mean()
avg_math_pass_spend
```

```python
avg_read_pass_spend = spending_grouped.groupby('Spending Ranges (Per Student)')['% Passing Reading'].mean()
avg_read_pass_spend
```

```python
overall_pass_spend = spending_grouped.groupby('Spending Ranges (Per Student)')['% Overall Passing'].mean()
overall_pass_spend
```

```python
scores_spending = pd.DataFrame([average_math_spending, average_read_spending, avg_math_pass_spend, avg_read_pass_spend, overall_pass_spend])
scores_spending
```

```python
spending_summary = scores_spending.transpose()
spending_summary
```
